Notebook showing how to write miner data to SQLite db and load db into pandas dataframe

```python
import asyncio
from pyasic import get_miner
from pyasic_dashboard.db import (
    load_db,
    write_data,
)
```

```python
# Update IP for your miner
ip = "192.168.0.164"
db_file = "write_data.db"
miner = await get_miner(ip)

# Start task to write data every minute
task = asyncio.create_task(
    write_data(
        miner=miner,
        data_file=db_file, 
        sleep_mins=1,
    )
)
```

```python
task.cancel()
try:
    await task
except asyncio.CancelledError:
    print("Stopped write_data task")
```

```python
# load SQLite db into pandas dataframe
df_db = load_db(db_file=db_file, table_name="data")
df_db.tail()
```
